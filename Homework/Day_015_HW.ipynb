```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.random((1000,3))
plot_data = pd.DataFrame(matrix,columns=['A','B','C'])
                                    
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)

plot_data['choice'] = indice

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'choice', 
                    vars = [x for x in list(plot_data.columns) if x != 'choice'])

# """
# Your Code Here
# Please replace "..." to correct plot function
# """
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot )
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r,kind="kde")

plt.show()
```

```python
sns.jointplot(x="A", y="C", data=plot_data, kind="kde")
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
scales = [200,150,300]
loc = [1000, 800, 600]
matrix2 = np.random.normal(loc=loc, scale=scales, size=(1000, 3))
plot_data2 = pd.DataFrame(matrix2,columns=['D','E','F'])

# 隨機給予 0, 1, 2 三種標籤
indice2 = np.random.choice([0,1,2], size=nrow)
plot_data2['choice'] = indice2

# 繪製 seborn 進階 Heatmap
grid2 = sns.PairGrid(data = plot_data2, size = 3, diag_sharey=False, 
                     hue = 'choice', 
                     vars = [x for x in list(plot_data2.columns) if x != 'choice'])

# # """
# # Your Code Here
# # Please replace "..." to correct plot function
# # """
grid2.map_upper(plt.scatter , alpha = 0.2)
grid2.map_diag(sns.kdeplot )
grid2.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r,kind="kde")

plt.show()
```
