畫出任意欄位的直方圖

```python
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')
pd.set_option('float_format','{:f}'.format)


import seaborn as sns
sns.distplot(app_train.AMT_INCOME_TOTAL, bins=200, hist=True,kde=False)
```

```python
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')
pd.set_option('float_format','{:f}'.format)

import seaborn as sns
sns.distplot(app_train[app_train.AMT_INCOME_TOTAL<2500000].AMT_INCOME_TOTAL, bins=200, hist=True,kde=False)
```

```python
# According to the plot, the distribution of income is skew right. Most of income are below $500,000.
# Use log transform to observe it. 
```

```python
app_train['log_AMT_INCOME_TOTAL'] = np.log10(app_train.AMT_INCOME_TOTAL)
```

```python
plt.ticklabel_format(style='plain',axis='x')
plt.ticklabel_format(style='plain',axis='y')
pd.set_option('float_format','{:f}'.format)


import seaborn as sns
sns.distplot(app_train.log_AMT_INCOME_TOTAL, bins=200, hist=True,kde=False)
```
